```python
data = pd.read_csv('artifact\data_preprocessing\preprocessed_data.csv')
data.head()
```

```python
X = data.drop(columns=['Price'])
y = data['Price']
```

```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# Split data into training and testing sets (80-20 split)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Dictionary of regression models to test
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'ElasticNet Regression': ElasticNet(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(),
    'SVR': SVR()
}
```

```python
# Create an empty list to store evaluation metrics
metrics_list = []

# Loop through models and calculate evaluation metrics
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    # Append results to the list
    metrics_list.append({
        'Model': model_name,
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'R^2': r2
    })

# Convert the list of results into a DataFrame
metrics = pd.DataFrame(metrics_list)

metrics
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import openpyxl

class RegressionModels:

    def __init__(self):
        self.data = pd.read_csv('artifact\data_preprocessing\preprocessed_data.csv')
        self.target_column = self.data['Price']
        os.makedirs('artifact\\regression_models', exist_ok=True)

    def regression_models(self):
        # Separate target variable and features
        X = self.data.drop(columns=[self.target_column])
        y = self.data[self.target_column]
        
        # Standardize the data
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)

        # Split data into training and testing sets (80-20 split)
        X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

        # Dictionary of regression models to test
        models = {
            'Linear Regression': LinearRegression(),
            'Ridge Regression': Ridge(),
            'Lasso Regression': Lasso(),
            'ElasticNet Regression': ElasticNet(),
            'Decision Tree': DecisionTreeRegressor(),
            'Random Forest': RandomForestRegressor(),
            'Gradient Boosting': GradientBoostingRegressor(),
            'SVR': SVR()
        }

        # Create a DataFrame to store evaluation metrics
        metrics = pd.DataFrame(columns=['Model', 'MAE', 'MSE', 'RMSE', 'R^2'])

        # Loop through models and calculate evaluation metrics
        for model_name, model in models.items():
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            
            mae = mean_absolute_error(y_test, y_pred)
            mse = mean_squared_error(y_test, y_pred)
            rmse = np.sqrt(mse)
            r2 = r2_score(y_test, y_pred)
            
            # Append results to the metrics DataFrame
            metrics = metrics.append({'Model': model_name, 'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R^2': r2}, ignore_index=True)

        # Save the evaluation metrics to an Excel file
        metrics.to_excel('artifact\\regression_models\\regression_metrics.xlsx', index=False)

        return metrics


if __name__ == '__main__':
    regression = RegressionModels()
    metrics = regression.regression_models()
    print(metrics)

    print('Regression models completed!')
```
