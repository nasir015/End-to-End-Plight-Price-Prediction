```python
data = pd.read_csv('artifact\data_cleaning\data_cleaned.csv')
data.head(5)
```

```python
# extract the year from the date column
data['Year_of_Journey'] = pd.DatetimeIndex(data['Date_of_Journey']).year
data['Month_of_Journey'] = pd.DatetimeIndex(data['Date_of_Journey']).month
data['Day_of_Journey'] = pd.DatetimeIndex(data['Date_of_Journey']).day
```

```python
data.head(5)
```

```python
# Define function to categorize season
def categorize_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Apply function to create 'Season' column
data['Season'] = data['Month_of_Journey'].apply(categorize_season)
```

```python
data.head(2)
```

```python
# Define function to categorize time of day
def categorize_time_of_day(dep_time):
    hour = int(dep_time.split(':')[0])
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    elif 18 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

# Apply function to create 'Time_of_Day' column
data['Time_of_Day'] = data['Dep_Time'].apply(categorize_time_of_day)
```

```python
# Define function to categorize flight duration
def categorize_flight_duration(duration):
    if duration <= 300:
        return 'Short'
    elif 300 < duration <= 600:
        return 'Medium'
    else:
        return 'Long'

# Apply function to create 'Flight_Duration_Category' column
data['Flight_Duration_Category'] = data['Duration(minute)'].apply(categorize_flight_duration)
```

```python
data.head(2)
```

```python
# Extract hours from Dep_Time and Arrival_Time columns
data['Dep_Hour'] = pd.to_datetime(data['Dep_Time'], format='%H:%M').dt.hour
data['Arr_Hour'] = pd.to_datetime(data['Arrival_Time'], format='%H:%M').dt.hour
```

```python
# Extract minutes from Dep_Time and Arrival_Time columns
data['Dep_Minute'] = pd.to_datetime(data['Dep_Time'], format='%H:%M').dt.minute
data['Arr_Minute'] = pd.to_datetime(data['Arrival_Time'], format='%H:%M').dt.minute
```

```python
# Drop the original Dep_Time and Arrival_Time columns
data.drop(columns=['Dep_Time', 'Arrival_Time'], inplace=True)
```

```python
data.head(2)
```

```python
# Calculate Price per minute of flight duration
data['Price_per_minute'] = data['Price'] / data['Duration(minute)']
```

```python
# Drop the temporary datetime columns
data.drop(columns=['Dep_Time_dt', 'Arr_Time_dt'], inplace=True)
```

```python
# Fill missing values in 'Route' column with an empty string
data['Route'] = data['Route'].fillna('')

# Create a new column for the total stops based on the Route column
data['Total_Stops'] = data['Route'].apply(lambda x: x.count('->') - 1 if 'non-stop' not in x else 0)
```

```python
data.head(2)
```

```python
# Calculate the difference in hours between departure and arrival
data['Journey_Hour_Diff'] = data['Journey_Duration'] / 60  # Convert minutes to hours
```
